# Calculationg pi using Monte Carlo methodology

## Relevant formulas
 - Area of a square: $A_s=(2r)^2$
 - Area of a circle: $A_c=\pi r^2$
Through using the ration of these areas, definition of $\pi$ can be obtained as $$\frac{A_c}{A_s} = \frac{\pi r^2}{4 r^2}=\frac{\pi}{4}$$
$$\rightarrow \pi=4\frac{A_c}{A_s}$$

## Visualization of the example 
![Darts](https://raw.githubusercontent.com/coderefinery/jupyter/main/example/darts.svg)

```python
import random
import matplotlib.pyplot as plt
```

```python
# Initializing number of throws to simulate pi
n_points = int(1e3)
```

```python
# Throw all the darts to the board
points = []
hits = 0

for _ in range(n_points):
    x,y = random.random(),random.random()
    if x*x+y*y<1.:
        hits += 1
        points.append((x,y,'red'))
    else:
        points.append((x,y,'blue'))
```

```python
# unzip coordinates and colors
x,y,c = zip(*points)

# Create a figure
fig, ax = plt.subplots(figsize=(6,6))

# Plot results
ax.scatter(x,y,c=c)
```

```python
# Compute estimate for pi and print it

frac = hits/n_points
f'{4*frac:.8f}'
```
